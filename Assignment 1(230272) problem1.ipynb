```python
import pandas as pd
import yfinance as yf
```

```python
# m 2023-04-01 to 2024-03-31
data=yf.download("META",start="2023-04-01",end="2024-03-31")
```

```python
data
```

```python
after=[]
before=[]
y=100000
for x in range(249):
    before.append(y)
    y=(data["Close"][x]*y)/data["Open"][x]
    after.append(y)
```

```python
# price on starting and end of the day
data["bef"]=before
data['aft']=after
```

```python
# ans(b)
returns=[]
for x in range(249):
    y=(data["Close"][x]-data["Open"][x])/data["Open"][x]
    returns.append(y*100)
```

```python
data['daily_ret']=returns
```

```python
# now calculation for nifty 50 index
# ans(c)
index_df=yf.download("^NSEI",start="2023-04-01",end="2024-03-31")
```

```python
index_df
```

```python
after=[]
before=[]
y=100000
for x in range(243):
    before.append(y)
    y=(index_df["Close"][x]*y)/index_df["Open"][x]
    after.append(y)
```

```python
index_df["after"]=after
index_df["before"]=before
```

```python
index_df
```

```python
returns=[]
for x in range(243):
    y=(index_df["Close"][x]-index_df["Open"][x])/index_df["Open"][x]
    returns.append(y*100)
```

```python
index_df['daily_ret']=returns
```

```python
index_df
```

```python
# assuming 0 returns in fd
Sharpe=index_df.daily_ret.mean()/index_df.daily_ret.std()
print("sharpe:",Sharpe)
```

```python
# ans(d)
plt.plot(data.index,data["aft"],label="stock")
plt.plot(index_df.index,index_df["after"],label="Nifty")
plt.legend()
plt.show()

```
