```python
s_date=datetime.datetime(2023,3,20)
e_date=datetime.datetime(2024,3,31)
```

```python
data=yf.download("META",start=s_date,end=e_date)
data.head(11)
```

```python
sum=0
for x in range(10):
    sum+=data["Close"][x]
```

```python
sum=sum/10
sum
```

```python
sma=[]
for x in  range(259):
    if x<10 :sma.append(sum)
    else:
        sum+=(data["Close"][x]-data["Close"][x-10])/10
        sma.append(sum)
```

```python
sma
```

```python
data["sma"]=sma
```
